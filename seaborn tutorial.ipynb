```python
#importing a preloded dataset from seaborn
import seaborn as sns
df= sns.load_dataset("iris")
df
```

```python

#line plot

sns.lineplot(x="sepal_length",y="petal_length",data=df)
```

```python
#barplot

sns.barplot(x="sepal_length",y="petal_length",data=df,hue="species")
```

```python
# scatter plot
sns.scatterplot(x="sepal_length",y="petal_width",data=df,hue="species")
```

```python
# histogram or Distribution plot
sns.distplot(df["sepal_length"],bins=20,kde=True,color="red")
```

```python
sns.distplot(df["sepal_length"],bins=20,kde=True,color="red",vertical=True)
```

```python
#joint plot

sns.jointplot(x="sepal_length",y="petal_width",data=df, kind="reg")
```

```python
# box plot

sns.boxplot(x="sepal_length",y="petal_width",data=df)
```
